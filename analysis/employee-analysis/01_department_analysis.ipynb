```python
# How is the perfomance distribution of the employees in the company?
df['PerformanceScore'].value_counts()
```

**Business Interpretation:**
- 78% of employees "Fully Meet" expectations (243/311 employees)
- 12% "Exceed" expectations - this could indicate limited growth opportunities

**Assessment:** Overall performance distribution appears healthy for a 311-employee organization.

**Next Question:** Are performance issues evenly distributed across departments, or are certain departments struggling more than others?

```python
# How is the perfomance distribution by department?
# Performance by department
df.groupby('Department')['PerformanceScore'].value_counts()
```

**Key Finding:** Production department shows disproportionate performance issues in raw numbers.

**The numbers:**
- Production: 23 underperformers (15 + 8)
- Sales: 5 underperformers (4 + 1)  
- IT/IS: 2 underperformers (1 + 1)

**Wait - Critical Analysis Question:** Are we comparing departments fairly? Production has 23 underperformers, but it also has way more employees than other departments. We need to calculate underperformance **rates**, not just raw counts.

**Next Analysis:** Let's calculate what percentage of each department is underperforming to get a true comparison.

```python
# First, get total employees by department
dept_totals = df.groupby('Department').size()
print("Total employees by department:")
print(dept_totals)

# Calculate underperformance rates
underperformers = df[df['PerformanceScore'].isin(['PIP', 'Needs Improvement'])]
underperformer_counts = underperformers.groupby('Department').size()

# Calculate rates
underperformance_rates = (underperformer_counts / dept_totals * 100).round(1)
print("\nUnderperformance rates by department:")
print(underperformance_rates)
```

**Critical Business Finding:** We have two distinct problems requiring different solutions.

**Sales Department Crisis:**
- 16.1% underperformance rate (1 in 6 employees)
- Small team (31 people) but direct revenue impact
- 5 underperformers potentially affecting company sales

**Production Department Challenge:**
- 11% underperformance rate (1 in 9 employees)  
- Large team (209 people) = 23 total underperformers
- Scale makes this a significant operational issue

**Strategic Implication:** Sales needs immediate intensive intervention due to high rate and revenue impact. Production needs systematic improvement due to scale.

## Deep-Dive: Sales Management Analysis

**Business Question:** Are Sales performance issues concentrated under specific managers, indicating management effectiveness problems?

**Hypothesis:** If Sales has culture/management issues, we'd expect to see underperformers clustered under certain managers.

```python
# Sales department only
sales_dept = df[df['Department'] == 'Sales']

# Performance by manager in Sales
sales_performance = sales_dept.groupby('ManagerName')['PerformanceScore'].value_counts()
print("Sales Performance by Manager:")
print(sales_performance)

# Manager team sizes in Sales
sales_manager_sizes = sales_dept.groupby('ManagerName').size()
print("\nSales Team Sizes by Manager:")
print(sales_manager_sizes)
```

```python
# Calculate underperformance rates by manager
sales_underperformers = sales_dept[sales_dept['PerformanceScore'].isin(['PIP', 'Needs Improvement'])]
sales_underperformer_counts = sales_underperformers.groupby('ManagerName').size()

# Calculate rates
sales_underperformance_rates = (sales_underperformer_counts / sales_manager_sizes * 100).round(1)
print("Sales Underperformance Rates by Manager:")
print(sales_underperformance_rates)
```

**Sales Management Analysis Results:**

**Underperformance rates by manager:**
- **Debra Houlihan:** 33.3% (1 of 3 employees) - Small team, high rate
- **John Smith:** 14.3% (2 of 14 employees) - Large team, moderate rate  
- **Lynn Daneault:** 15.4% (2 of 13 employees) - Large team, moderate rate
- **Janet King:** 0% (0 of 1 employee) - Single employee, performing well

**Critical Business Insight:** While Debra Houlihan shows a high rate, her team is only 3 people (small sample). The two main managers (John Smith and Lynn Daneault) show similar underperformance rates (~14-15%).

**Conclusion:** Sales performance issues are **NOT management-driven** but appear to be **systemic to the sales function itself.**

**Root Cause Assessment:** The evidence points to:
- High-stress nature of sales work affecting performance
- Possible skills/training gaps in sales methodology
- Market or territory challenges
- Inadequate support systems for sales teams

**Strategic Implication:** Focus on sales training, process improvement, and stress management rather than management changes.
