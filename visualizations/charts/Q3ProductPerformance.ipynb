```python
import matplotlib.pyplot as plt
import numpy as np

# Your business data
categories = ['Electronics', 'Clothing', 'Books', 'Home']
actual = [450, 380, 120, 290]  # in thousands
targets = [500, 300, 150, 280]
last_year = [420, 350, 110, 260]
```

```python
colors = []
for i in range(len(actual)):
    if actual[i] >= targets[i]:
        colors.append('green')
    else:
        colors.append('red')
colors
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Gray background bars for targets
ax.bar(categories, targets, color='lightgray', alpha=0.7, label='Target')

# Your colored bars for actual performance
ax.bar(categories, actual, color=colors, label='Actual')

# show lines on top so the green doesn't cover the target where green is higher than grey
#ax.plot(categories, targets, color='black', marker='_', linestyle='None', label='Target')
# or
# For each category position, draw a line spanning the bar width
for i, target in enumerate(targets):
    ax.hlines(y=target, xmin=i-0.4, xmax=i+0.4, color='black', linewidth=0.5)
```

```python
from matplotlib.patches import Patch

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Q3 Product Category Performance vs Targets')
ax.set_ylabel('Revenue ($000s)')

for i in range(len(categories)):
    if actual[i] >= targets[i]:
        # Draw green base + gold top
        ax.bar(categories[i], targets[i], color='green')
        ax.bar(categories[i], actual[i] - targets[i], color='gold', bottom = targets[i], hatch='*', edgecolor='darkgoldenrod')
    else:
        # Draw red bar + target line  
        ax.bar(categories[i], actual[i], color='red')
        ax.bar(categories[i], targets[i] - actual[i], color='lightgray', alpha=0.7, bottom=actual[i], hatch='/////')
        # or
        ax.hlines(y=targets[i], xmin=i-0.4, xmax=i+0.4, color='black', linewidth=2)

# Add invisible bars just for legend
legend_elements = [
    Patch(facecolor='green', label='Meeting Target'),
    Patch(facecolor='gold', hatch='*', edgecolor='darkgoldenrod', label='Exceeding Target'),
    Patch(facecolor='red', label='Below Target'),
    Patch(facecolor='lightgray', alpha=0.7, hatch='/////', label='Performance Gap')
]

ax.legend(handles=legend_elements)
```
